```python
import numpy as np
import scipy.signal as sg
```

```python
## LE CONTENU DE CETTE CELLLULE EST A SOUMETTRE SUR INGINIOUS
def filtre(n1,d1,n2,d2,y1,x2,fe):
    """
    Filtre de bruit parasites
    Args:
        n1 (numpy array): coefficients du numérateur du filtre 1
        d1 (numpy array): coefficients du dénominateur du filtre 1
        n2 (numpy array): coefficients du numérateur du filtre 2
        d2 (numpy array): coefficients du dénominateur du filtre 2
        y1 (numpy array): Le son à filtrer
        x2 (numpy array): Le son parasite
        fe (int): Fréquence d'échantillonage
    Returns:
        yest (numpy array): Le son filtré
        H3 (numpy array): La fonction de transfert du filtre appliqué

    """
    H1 = sg.TransferFunction(n1,d1,dt=1/fe)
    H2 = sg.TransferFunction(n2,d2,dt=1/fe)
    h1 = H1.to_zpk()
    h2 = H2.to_zpk()
    z3 = np.append(h2.poles,h1.zeros)
    p3 = np.append(h1.poles,h2.zeros)
    k3 = h1.gain/h2.gain
    H3 = sg.ZerosPolesGain(z3,p3,k3,dt=1/fe).to_tf()
    x3 = sg.lfilter(H3.num,H3.den,x2)
    yest = y1 - x3

    return (H3,yest)
```
